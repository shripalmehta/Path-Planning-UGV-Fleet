```python
import geopandas as gpd
```

```python
# Paths to the Shapefiles
road_link_file = 'data/oproad_esri_gb/data/HP_RoadLink.shp'
road_node_file = 'data/oproad_esri_gb/data/HP_RoadNode.shp'

# Load the Shapefiles
road_links_gdf = gpd.read_file(road_link_file)
road_nodes_gdf = gpd.read_file(road_node_file)
```

```python
# Display the first few rows of the road links
road_links_gdf.head()
```

```python
# Get information about the GeoDataFrame
road_nodes_gdf.info()
```

```python
# Display the first few rows of the road nodes
road_nodes_gdf.head()
```

```python
# Get information about the road links GeoDataFrame
road_links_gdf.info()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))

# Plot the road links
road_links_gdf.plot(ax=ax, color='blue')

# Plot the road nodes
road_nodes_gdf.plot(ax=ax, color='red')

plt.show()

fig.savefig('plot.png')
```
